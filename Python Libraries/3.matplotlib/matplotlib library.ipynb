```python
plt.figure() # رسم یک شکل نمودار
# رسم اولین پنل 
plt.subplot(2, 1, 1) # (شماره پنل ها، تعداد ستون ها، تعداد سطرها)
plt.plot(x, np.sin(x))
# ساخت دومین پنل
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

```python
plt.figure() # رسم یک شکل نمودار
# رسم اولین پنل 
plt.subplot(2, 2, 1) # (شماره پنل ها، تعداد ستون ها، تعداد سطرها)
plt.plot(x, np.sin(x))
# ساخت دومین پنل
plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x));

# رسم سومین پنل 
plt.subplot(2, 2, 1) # (شماره پنل ها، تعداد ستون ها، تعداد سطرها)
plt.plot(x, np.tan(x))
# ساخت چهارمین پنل
plt.subplot(2, 2, 2)
plt.plot(x, np.exp(x));
```

```python
plt.figure() # رسم یک شکل نمودار
# رسم اولین پنل 
plt.subplot(2, 2, 1) # (شماره پنل ها، تعداد ستون ها، تعداد سطرها)
plt.plot(x, np.sin(x))
# ساخت دومین پنل
plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x));

# رسم سومین پنل 
plt.subplot(2, 2, 3) # (شماره پنل ها، تعداد ستون ها، تعداد سطرها)
plt.plot(x, np.tan(x))
# ساخت چهارمین پنل
plt.subplot(2, 2, 4)
plt.plot(x, np.exp(x));
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)') # with - and green
plt.plot(x, np.cos(x), ':b', label='cos(x)') # with : and blue
plt.axis('equal')
plt.legend(); # fiqure's guide
```

```python
rng = np.random.RandomState() #random distribution
for marker in ['p', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,label="marker='{0}'".format(marker))
    plt.legend(numpoints=1)
    plt.xlim(0, 1.8);
    plt.ylim(0, 1.2);
```

```python
x = np.linspace(0, 10, 35)
y = np.sin(x)
plt.plot(x, y, 'o', color='black');
```

```python
plt.plot(x, y, '-ok'); # line (-), circle marker (o), black (k)
```

```python
plt.plot(x, y, '-p', color='gray',
         markersize=15, linewidth=1,
         markerfacecolor='red',
         markeredgecolor='gray',
         markeredgewidth=5)
plt.ylim(-1.2, 1.2);
```
